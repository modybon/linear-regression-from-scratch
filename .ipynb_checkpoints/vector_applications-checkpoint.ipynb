# K-Means

```python
def k_means(x,k=3, n = 100):
    centroids_idx = [np.random.randint(0, len(x)) for i in range(k)]
    centroids = x[centroids_idx]
    labels = np.zeros(len(x))
    for _ in range(n):
        for i in range(len(x)):
            distances = np.linalg.norm(centroids-x[i], axis=1)
            closest_centroid = np.argmin(distances)
            labels[i] = closest_centroid

        for c in range(k):
            cen_idx = np.where(labels==c)[0]
            cen_data = x[cen_idx]
            centroids[c] = np.mean(cen_data, axis=0)
    
    labeled_data = np.concatenate((x, labels.reshape(-1, 1)), axis=1)
    columns = [f"c_{i}" for i in range(x.shape[1])] + ['label']
    labeled_data_df = pd.DataFrame(labeled_data, columns=columns)
    return centroids, labeled_data_df
```

```python
data = np.random.randn(1000,2)
centroids, df = k_means(data,k=4,n=100)
sns.scatterplot(data=df, x='c_0', y='c_1', hue='label')
```

```python
n = 5
fig, axes = plt.subplots(1, n, figsize=(15,5))

for i in range(1,n+1):
    centroids, df = k_means(data, n=n*10)
    sns.scatterplot(data=df, x='c_0', y='c_1', hue='label', ax=axes[i-1])
    axes[i-1].set_title(f"n = {i*10}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create kernel and signal
kernel = np.array([1, -1])
signals = np.random.randn(100)
#signals[10:20] = 1  # Inject pattern

# Original signal DataFrame
signals_df = pd.DataFrame({'ss': signals})
signals_df['label'] = 'signal'

# Apply kernel as 1D convolution-like operation
feature_map = np.zeros_like(signals)
for t in range(1, len(signals)):
    feature_map[t] = np.dot(kernel, signals[t-1:t+1])  # shape (2,)

# Filtered signal DataFrame
feature_map_df = pd.DataFrame({'ss': feature_map})
feature_map_df['label'] = 'filtered signal'

# Concatenate for plotting
signal_filtered_df = pd.concat([signals_df, feature_map_df], axis=0, ignore_index=True)
signal_filtered_df['x'] = np.tile(np.arange(len(signals)), 2)  # 2x 100

# Plot
_, axes = plt.subplots(1, 3, figsize=(14, 4))

# Plot kernel
axes[0].plot(kernel, 'ks-')
axes[0].set_title('Kernel')
axes[0].set_xlim([-1, 3])

# Plot original signal
axes[1].plot(signals, 'ks-')
axes[1].set_title('Signal')
axes[1].set_xlim([0, 30])

# Plot both signals
sns.lineplot(data=signal_filtered_df, x='x', y='ss', hue='label', ax=axes[2])
axes[2].set_title('Signal + Feature Map')
axes[2].set_xlim([0, 30])

plt.tight_layout()
plt.show()
```

```python
kernel = np.array([0,.1,.3,.8,1,.3,.1,0])
kernel = kernel / np.sum(kernel)

Nkernel = len(kernel)
halfkernel = Nkernel//2

Nsignal = 100
timeseries = np.random.randn(Nsignal)

_, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].plot(kernel, 'ks-')
ax[0].set_title('kernel')
ax[0].set_xlim([-1,Nsignal])

ax[1].plot(timeseries, 'ks-')
ax[1].set_title('Time series signal')
```

```python
filtsig = timeseries.copy()

for t in range(halfkernel+1, Nsignal-halfkernel):
    filtsig[t] = np.dot(kernel, timeseries[t-halfkernel: t+halfkernel])

_, ax = plt.subplots(1,2, figsize=(12,4))
ax[0].plot(kernel, 'ks-')
ax[0].set_title('kernel')
ax[0].set_xlim([-1,Nsignal])

ax[1].plot(timeseries, color='k', label='Original', linewidth=1)
ax[1].plot(filtsig, '--', color=[.6,.6,.6], label='Smoothed', linewidth=2)
ax[1].set_title('Time series signal')
plt.legend()
```
